```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("./Coleccionista.csv")
```

```python
data.shape
```

```python
data.head()
```

## Variables independientes

```python
X = data[['Bidders','Age']]
X
```

## Variable dependiente

```python
y = data[['Price']]
y
```

```python
model.intercept_
```

```python
#Ajustar el modelo de regresion multiple
model = LinearRegression()
model.fit(X,y) #entrenando el modelo
y_pred = model.predict(X)
#y_pred 

#Metricas estadisticas
mse = mean_squared_error(y, y_pred) #Error de la media cuadratica
r2 = r2_score(y, y_pred) #varianza
```

```python
model.intercept_
```

```python
print('Coeficientes: ', model.coef_)
print('Error cuadratico de la media: ', mse)
print('Varianza: ', r2)
```
