```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("ConstructorInmobiliario.csv")
df.head()
```

```python
x = df[["Income","Size","Senior","Education"]]
x
```

```python
y = df[["SqrFt"]]
y
```

```python
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)


mse = mean_squared_error(y,y_pred)
r2 = r2_score(y,y_pred)
```

```python
model.intercept_
```

```python
print("Coeficientes son:", model.coef_)
print("Error cuadratico: ", mse)
print("Varianza", r2)
```

```python
Y = model.predict([[60.8,2,0,4]])
print("La relacion es:", Y)
```
