# 3. Cincinnati Paint Company

¿Ganan más en comisiones por ventas los agentes que recorren más millas y hacen más llamadas?

Variables: 

- Cantidad que ganaron por comisiones el mes anterior (y)
- Numero de millas que recorrieron durante ese mes (x_1)
- Numero de llamadas de ventas del periodo (x_2)

¿Hay una interacción significativa entre el numero de llamadas de ventas y las millas recorridas?

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("CincinnatiPaintCompany.csv")
df.head()
```

```python
# Determinar variables independientes
x = df[["Calls", "Driven"]]
x
```

```python
# Determinar variable dependiente
y = df[["Commissions"]]
y
```

```python
# Ajustando el modelo para encontrar el modelo de regresion múltiple
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

# Error de la media cuadrática
mse = mean_squared_error(y, y_pred)

# Varianza
r2 = r2_score(y, y_pred)

print("Error de la media cuadrática: ", mse)
print("Varianza: ", r2)
```

```python
# Valor de intersección
model.intercept_
```

```python
# Coeficientes
print("Coeficientes son:", model.coef_)
print("Error cuadratico: ", mse)
print("Varianza", r2)
```

```python
# ¿Ganan más en comisiones por ventas los agentes que recorren más millas y hacen más llamadas?
Y = model.predict([[149, 2776]])
print("Ganancia estimada de comisión según llamadas y millas:", Y)
```
