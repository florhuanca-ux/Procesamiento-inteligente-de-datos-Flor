```python
import random

class Individuo:
    def __init__(self, especie):
        self.especie = especie
        self.adn = [random.randint(0, 1) for _ in range(5)]
        self.energia = 100
        print(f"Un nuevo {self.especie} ha nacido.")

    def alimentarse(self):
        self.energia += 20
        print(f"El individuo comió. Energía actual: {self.energia}")

    def desplazarse(self):
        # En AG, desplazarse es explorar el espacio de búsqueda
        consumo = random.randint(5, 15)
        self.energia -= consumo
        print(f"El individuo se desplazó y consumió {consumo} de energía.")

    def envejecer(self):
        self.energia -= 10
        if self.energia <= 0:
            print("El individuo ha muerto por vejez.")

class Poblacion:
    def __init__(self, nombre_ecosistema):
        self.nombre = nombre_ecosistema
        self.habitantes = []
        self.clima = "Estable"

    def colonizar(self, cantidad):
        for _ in range(cantidad):
            nuevo = Individuo("Organismo")
            self.habitantes.append(nuevo)
        print(f"Población {self.nombre} ahora tiene {len(self.habitantes)} habitantes.")

    def cambiar_clima(self, nuevo_clima):
        self.clima = nuevo_clima
        print(f"El clima ha cambiado a: {self.clima}")

    def ciclo_de_vida(self):
        print("\n--- Iniciando ciclo de vida ---")
        for h in self.habitantes:
            h.desplazarse()
            h.envejecer()

    def mostrar_censo(self):
        print(f"\nCenso en {self.nombre}:")
        for i, h in enumerate(self.habitantes):
            estado = "Vivo" if h.energia > 0 else "Muerto"
            print(f"Sujeto {i}: {estado} | Energía: {h.energia}")

# --- Prueba rápida ---
bosque = Poblacion("Reserva Natural")
bosque.colonizar(3)
bosque.ciclo_de_vida()
bosque.mostrar_censo()
```

#### La Naturaleza como Optimizadora
De acuerdo con Kuri (2002), la naturaleza no busca una perfección absoluta de forma inmediata, sino que opera bajo un esquema de adaptación. Los sistemas biológicos resuelven problemas complejos (supervivencia) mediante procesos de optimización que han tomado millones de años. En el cómputo evolutivo, imitamos esta "búsqueda" para encontrar soluciones óptimas en espacios de datos complejos.

#### Un poco de Biología
Para programar estos algoritmos, abstraemos conceptos biológicos:
* **Individuo:** Es la unidad básica (una solución potencial).
* **Genotipo/ADN:** La información codificada que define al individuo.
* **Fenotipo:** La expresión de esos genes (sus acciones o características).
* **Selección:** El proceso donde solo los más aptos (con más energía o mejor "fitness") permanecen.
